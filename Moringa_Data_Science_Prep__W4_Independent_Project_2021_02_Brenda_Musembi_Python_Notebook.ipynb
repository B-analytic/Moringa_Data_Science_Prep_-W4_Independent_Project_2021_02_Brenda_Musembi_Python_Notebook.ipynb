```
#importing pandas and numpy libraries
#
import pandas as pd
 
import numpy as np
```

**1.1.LOADING DATASETS**

```
#Loading dataset
#
url="http://bit.ly/autolib_dataset"
df=pd.read_csv(url)
```

```
#previewing data set
df.head(5)
```

**2.CLEANING DATASET.**

**2.1.Uniformity**

```
#changing date to datetime format
#
df['Date']=pd.to_datetime(df[['month','day','year']])
df
```

**2.2.Validity.**

```
#dropping irrelevant columns
#
df.drop(["Displayed comment","ID","Geo point","Scheduled at","Charge Slots","Charging Status","year","month","day"],axis=1,inplace=True)
df.head(5)
```

```
#renaming columns
#
df.columns=df.columns.str.replace(" ","_")
df.head(5)
```

```
#changing values in car column
#
df['Cars']=df['Bluecar_counter']+df['Utilib_counter']+df['Utilib_1.4_counter']
df=df.rename(columns={'Cars':'Total_Cars'})
df.head(5)      
```

**2.3.Completeness**

```
#checking for null values
#
df.info()
```

**2.4.Consistency**

```
#checking for duplicates
#
df.duplicated()
```

```
#dropping any duplicate records
#
df.drop_duplicates(inplace=True)
df

```

**3.0.PRIORITY LISTS**

```
#dropping any records that Kind
#of station is a centre
#
df=df[df.Kind != 'CENTRE']
df
```

```
#dropping records status of the
#station is not operational
#
df=df[df.Rental_status == 'operational']
df
```

```
#dropping records with subscription 
#status is future
df=df[df.Subscription_status != 'future']
df
```

```
#checking how many record
#in the data frame priority list
#
df.info()
```

**4.0.DATA ANALYSIS**

```
#most popular station
#
df["sum_of_resources"]=df['Bluecar_counter']+df['Utilib_counter']+df['Utilib_1.4_counter']
popular_station=pd.pivot_table(df,values=["sum_of_resources"],index=["Public_name"],aggfunc=np.sum)

popular_station.sort_values(by="sum_of_resources",ascending=False).head(2)
```

```
#records in which car was picked
#in paris in april 2018
#
paris_renting=df[(df['City']=='Paris') & (df['Date']>='2018-04-01') &  (df['Date']<='2018-04-30')]
paris_renting


```

```
#creating pivot table to know
#hours for picking up blue cars
#
table1=pd.pivot_table(paris_renting,values=["Bluecar_counter","Utilib_counter","Utilib_1.4_counter"],index=['hour'],aggfunc=np.sum)
table1
```

```
#list for best hour
#for picking electic cars
#
sum=table1['Bluecar_counter']+table1['Utilib_counter']+table1['Utilib_1.4_counter']
table1['total_cars1']=sum
table1.sort_values(by='total_cars1',ascending=0).head(1)
```

```
#list for best hour
#for picking electic cars
#
hour_returning=pd.pivot_table(df,values=['Slots'],index=['hour'],aggfunc=np.sum)
hour_returning.sort_values(by='Slots',ascending=False).head(1)
```

```
#most popular postal code and station
#for blue cars
#
df.pivot_table(values=['Bluecar_counter'],index=['Postal_code','Public_name'],aggfunc=np.sum).sort_values(by='Bluecar_counter',ascending=False).head(1)
```
